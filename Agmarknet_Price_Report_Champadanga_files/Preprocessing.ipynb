```python
potato_data = pd.read_csv('C:\\Users\\bipas\\Work\\Potato_price_prediction\\Agmarknet_Price_Report_Champadanga.csv')
rainfall_data = pd.read_csv('C:\\Users\\bipas\\Work\\Potato_price_prediction\\Rain Data\\Rainfall_WB.csv')
```

```python
rainfall_data
```

```python
# Extract month and year from the Date column
potato_data['Price Date'] = pd.to_datetime(potato_data['Price Date'])
potato_data['Price Date'] = potato_data['Price Date'].dt.strftime('%Y-%m-%d')
potato_data['Date'] = pd.to_datetime(potato_data['Price Date'])



potato_data['Month'] = potato_data['Date'].dt.strftime('%b').str.upper()
potato_data['Year'] = potato_data['Date'].dt.year
```

```python
potato_data
```

```python
potato_data.loc[(potato_data['Year'] ==2009)]
```

```python
potato_data.drop(['Unnamed: 0','Sl no.','Price Date'], axis=1)
```

```python
# Merge the potato_data with rainfall_data on Month and Year
final_data = pd.merge(potato_data, rainfall_data, left_on=['Year'], right_on=['YEAR'], how='inner')
```

```python
final_data
```

```python
final_data.isnull().sum()
```

```python
# Extract the rainfall data for the corresponding month
final_data['Rainfall'] = final_data.apply(lambda row: row[row['Month']], axis=1)
```

```python
final_data
```

```python
# Select the required columns for the final output
output = final_data[['Date', 'Rainfall', 'Min Price (Rs./Quintal)','Max Price (Rs./Quintal)','Modal Price (Rs./Quintal)','JAN','FEB',
'MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','Month','Year']]
```

```python
# Select the required columns for the final output
final_output = final_data[['Date', 'Rainfall', 'Modal Price (Rs./Quintal)','Max Price (Rs./Quintal)']]
```

```python
# Save the final output to a CSV file
final_output.to_csv('final_price_champadanga_rainfall_data.csv', index=False)
```
